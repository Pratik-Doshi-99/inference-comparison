```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
req_metrics = pd.read_csv('output/sharegpt_test/metrics_20250123_091308.csv')
```

```python
req_metrics.head()
```

```python
req_metrics.dtypes
```

```python
num_columns = 3
fig, axes = plt.subplots(1, num_columns, figsize=(15, 8))
fig.suptitle('Client-side metrics')

for i, column in enumerate(['ttft', 'throughput', 'avg_inter_token']):
    sns.violinplot(y=req_metrics[column], ax=axes[i], inner="point", palette="muted")
    axes[i].set_title(f'{column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Values')

    # Adjust y-axis limits for better scaling
    axes[i].set_ylim([req_metrics[column].min() - (0.1 * np.array(req_metrics[column]).ptp()), 
                      req_metrics[column].max() + (0.1 * np.array(req_metrics[column]).ptp())])

plt.tight_layout()
plt.show()
```
